```python
def calH(alpha):
    alpha[0,0]-=np.linalg.norm(alpha)
    omega=alpha/np.linalg.norm(alpha)
    oo=omega@omega.T
    H=np.eye(oo.shape[0])-2*oo
    return H

def calQ(calH,A):
    n=A.shape[0]
    subA=A.copy()
    for ii in range(n-1):
        alpha=subA[:,0].reshape(-1,1)
        h=calH(alpha)
        H=np.eye(n)
        H[ii:,ii:]=h
        A=H@A
        subA=A[ii+1:,ii+1:].copy()
    return A[:,:]
B=np.array([[0,3,1],[0,4,-2],[2,1,1]])
A=np.array([[10,7,8,7],
            [7,5,6,5],
            [8,6,10,9],
            [7,5,9,10]])
R=calQ(calH,A)
print(R)
```
